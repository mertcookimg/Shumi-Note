# テイラーの定理

詳細：https://manabitimes.jp/math/1001

テイラーの定理は微分可能な$f(x)$を多項式に変形したいときに便利な定理です。

---

**テイラーの定理**

閉区間$[a, x]$で$n$回微分可能な関数$f$について、

$$
f(x) = \sum^{n-1}_{k=0} f^{(k)}(a) \frac{(x-a)^k}{k!} + f^{(n)}(c) \frac{(x-a)^n}{n!}
$$

を満たす$c (a < c < x)$ が存在します。1つ目の項が多項式、2つ目の項が$f$との誤差になります。

---

例として $f(x) = \exp(x)$を$n=3$、$[2, x]$の間にテイラーの定理を適用すると、

$$f(x) = \exp(2) + \exp(2)(x - 2) + \frac{\exp(2)}{2} (x - 2)^2 + \frac{\exp(c)}{6} (x - 2)^3$$

を満たす$c \in [2, x]$が存在します。

```python
import numpy as np
import matplotlib.pyplot as plt

a = 2
x = np.linspace(2.1, 5, 100)
f = np.exp(x)
f_taylor = np.exp(a) + np.exp(a) * (x - a) + np.exp(a) / 2 * (x - a) ** 2

plt.plot(x, f, label=r"$f=\exp(x)$")
plt.plot(x, f_taylor, label="Second Taylor")

# cを計算します。 c \in [a, x]をチェックします。
error = f - f_taylor  # テイラー展開の誤差
c = np.log(error * 6 / ((x - a) ** 3))
assert np.all(c <= x)

f_taylor_theorem = np.exp(a) + np.exp(a) * (x - a) + np.exp(a) / 2 * (x - a) ** 2 + np.exp(c) / 6 * (x - a) ** 3
plt.plot(x, f_taylor_theorem, label="Taylor Theorem")

plt.legend()
```

テイラーの定理は平均値の定理を一般化しています。また、テイラー展開の原理もテイラーの定理の拡張です。

テイラーの定理の証明には次のロルの定理を使います

---

**ロルの定理**

区間$[a, b]$で連続、$(a, b)$で微分可能、$f(a) = f(b)$ である関数$f(x)$に対して、
$a < c < b$なる$c$で$f'(c)=0$を満たす$c$が存在する。
