# コンパクト集合

参考
* [実数空間におけるコンパクト集合](https://wiis.info/math/real-number/topology/compact-set/)

実数空間の部分集合Aを覆う開集合族（開被覆）を任意に選んだとき、それに対して有限部分被覆が必ず存在する場合には、Aはコンパクト集合であると言います。
コンパクト集合はよく出てくる概念なのでちゃんとまとめておきましょう。

## 実数集合の被覆

実数空間$\mathbb{R}$の部分集合$A$が与えられたとき、これが$\mathbb{R}$の部分集合族$\left\{A_\lambda\right\}_{\lambda \in \Lambda}$の和集合の部分集合になる場合、すなわち

$$
A \subset \bigcup_{\lambda \in \Lambda} A_\lambda
$$

のとき、$\left\{A_\lambda\right\}_{\lambda \in \Lambda}$は$A$の被覆（covering）と呼びます。特に$\left\{A_\lambda\right\}_{\lambda \in \Lambda}$が有限だったり開集合だったりすると、それを有限被覆や開被覆と呼びます。

---

例えば$A=(0, 3)$について、$A_1 =(0,1), A_2 =\left(\frac{1}{2}, 2\right), A_3 =\left(\frac{3}{2}, \frac{7}{2}\right)$の和集合は$(0, 3)\in(0, 7/2)$を満たすので、これは$A$の有限被覆になります。

---

また、$M\subset \Lambda$が存在して、
$$A \subset \bigcup_{\lambda \in M} A_\lambda$$
が成り立つ場合、このような集合族$\left\{A_\lambda\right\}_{\lambda \in M}$を$\left\{A_\lambda\right\}_{\lambda \in \Lambda}$の部分被覆といいます。

---

**例**

* $A=(0, 1)$
* $\mathbb{R}$の部分集合$A_i=\left(-\frac{i}{2}, \frac{i}{2}\right)$

このとき、$(0,1) \subset \bigcup_{i \in \mathbb{N}} A_i$なので$\left\{A_i\right\}_{i \in \mathbb{N}}$は$A$の被覆です。ここで、$A_1 \cup A_2=\left(-\frac{1}{2}, \frac{1}{2}\right) \cup(-1,1) =(-1,1)$なので、$\{A_1, A_2\}$も$A$の被覆です。

このとき、$\{A_1, A_2\}$は$\left\{A_i\right\}_{i \in \mathbb{N}}$の有限部分被覆です。

---


## コンパクト集合

$\mathbb{R}$の部分集合$A$がコンパクト集合であることとは、$A \subset \bigcup_{\lambda \in \Lambda} A_\lambda$を満たす$\mathbb{R}$上の開集合からなる集合族$\left\{A_\lambda\right\}_{\lambda \in \Lambda}$を任意に選んだとき、それに対して有限個の添字$\lambda_1, \cdots, \lambda_n \in \Lambda$が存在して、
$$
A \subset \bigcup_{i=1}^n A_{\lambda_i}
$$
とできることを意味します。
つまり、$\mathbb{R}$上の開集合の族によって$A$を覆っても、その中の有限この開集合によって$A$が覆える場合に$A$はコンパクトであると呼びます。

---

**例**

$A=\{a_1, a_2, \dots, a_n\}\subset \mathbb{R}$ を実数空間の有限集合とする。
ここで、$\{A_\lambda\}_{\lambda \in \Lambda}$を$A\subset \bigcup_{\lambda \in \Lambda}A_\lambda$なる$A$の開被覆とする。

このとき、$\{a_1, a_2, \dots, a_n\}\subset\bigcup_{\lambda \in \Lambda}A_\lambda$なので、各$a_i$に対して、$a_i \in A_{\lambda_i}$であるような$\lambda_i \in \Lambda$が存在する。
よって$\{a_1, a_2, \dots, a_n\}\subset\bigcup_{\lambda_i \in \Lambda}^n A_\lambda$なる部分開被覆が存在する。

よって有限集合はコンパクトです。

**最大値・最小値の定理**

参考：[コンパクト性とは：有界閉集合、最大値の定理を例に](https://math-fun.net/20210324/12430/)

$I=[a, b]$を有界な閉区間とし、$f: I\to \mathbb{R}$を連続関数とします。このとき、$f$は$I$上で最大値、最小値を持ちます。

これはハイネ・ボレルの被覆定理を使えば簡単に証明できます。
定理より$I$はコンパクトであり、一般にコンパクト集合の連続写像による像はコンパクトになります。
よって$f(I)$はコンパクトであり、定理からこれは有界な閉集合です。

ここで、実数の公理から$\mathbb{R}$の有界な部分集合には上限$M$と下限$m$が存在し、$f(I)$は閉集合なので$M\in f(I)$であり、同様に$m\in f(I)$です。
$$
f(I)=\{y \mid \exists x \in I, f(x)=y\}
$$
なので、$f(x_1)=m$、$f(x_2)=M$となる$x_1$と$x_2$が存在します。
