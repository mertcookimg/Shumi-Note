# Danskinの定理

参考
* [Danskin's theorem](https://en.wikipedia.org/wiki/Danskin%27s_theorem)
* [What is Danskin’s Theorem?](https://statisticaloddsandends.wordpress.com/2022/11/10/what-is-danskins-theorem/)

Danskinの定理を使うと，次の形式の関数の凸性とその微分が簡単に示せます．

$$f(x)=\max _{z \in Z} \phi(x, z)$$

準備
* $\phi: \mathbb{R}^n \times Z \mapsto \mathbb{R}$：連続かつ任意の$z\in Z$で，$x$について凸関数
    * $Z \in \mathbb{R}^m$はコンパクト集合
* $Z_0(x)=\left\{\bar{z}: \phi(x, \bar{z})=\max _{z \in Z} \phi(x, z)\right\}$：各$x$で$z$について最大値を与えるポイント


このとき，次が成立します：

---

**Danskinの定理**

1. $f(x)$は凸関数
2. $Z_0(x)$が単一の要素$\bar{z}$からなるとします．つまり，$Z_0(x)=\{\bar{z}\}$とします．このとき，$f(x)$の微分は$\frac{\partial f}{\partial x}=\frac{\partial \phi(x, \bar{z})}{\partial x}$
3. 方向微分のやつもあるけど省略

---

例を見てみましょう．

$C\subseteq \mathbb{R}^n$を凸なコンパクト集合とします．
$\gamma \geq 0$を定数とします．このとき，次の関数$\phi: \mathbb{R}^n \times C \mapsto \mathbb{R}$を考えましょう．

$$
\phi(x, z)=-x^{\top} z-\frac{\gamma}{2} z^{\top} z
$$

これは明らかに連続です．また，固定された$z$について$\phi$は線形なので，凸関数です．
よってDanskinの定理の仮定を満たします．このとき

$$
f(x) =\max _{z \in C} \phi(x, z) =-\min _{z \in C}\left\{x^{\top} z+\frac{\gamma}{2} z^{\top} z\right\} .
$$

ここで，$\min$の内側について見てみましょう．これは強凸関数かつ凸集合について$\min$を取っているので，minimizerは一つだけ存在します．
よって，$f$の$x$での微分は

$$
\nabla f(x)=\frac{\partial \phi(x, \hat{z}(x))}{\partial x}=-\hat{z}(x)
$$

で与えられます．

```python
import numpy as np
import matplotlib.pyplot as plt

gamma = 0.1
C = np.linspace(-5, 5, 100)

def phi(x, z):
    return - x * z - gamma / 2 * z * z


def f(x):
    return phi(x, C).max()


X = np.linspace(-10, 10, 100)
res = []
for x in X:
    res.append(f(x))

plt.plot(X, res)
```
