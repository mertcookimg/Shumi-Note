# ベクトル空間

参考
* [線形代数汎論](https://amzn.asia/d/cQQdYH2): ３章

空でない集合$V$が体（四則演算が定義された集合）のベクトル空間であるとは、$V$の任意の元$x, y$に対して
* 和$x+y$が$V$の元として一意に定まり、
* $x$の$\lambda$倍が$V$の元として一意に定まり、
* 以下の演算を満たす
    * $x, y, z \in V$に対して $(x + y) + z= x + (y + z)$
    * $\cdots$ その他は教科書参照。


体$K$の上野ベクトル空間$V$の$m$本のベクトルの集合$X=\{x_1, x_2, \cdots, x_m\}$のスカラーの集合$\{\lambda_1, \cdots, \lambda_m\}$を係数とする線型結合とは、ベクトル
$\lambda_1 x_1 + \cdots + \lambda_m x_m$

のことである。
$X$が線形従属であるとは、すべてが$0$でないスカラー$\{\lambda_1, \cdots, \lambda_m\}$を係数とする線型結合で$0$となるものがあること、すなわち
$\lambda_1 x_1 + \cdots + \lambda_m x_m=0$
であることをいう。
線形従属でないならば線形独立であるという。


## ベクトル空間の次元と基底

ベクトル空間$V$の極大線形独立集合の要素数のことを$V$の次元という。
また、その集合のことを$V$の基底と呼ぶ。

$m\times n$型の行列$A$の$n$本の列ベクトルを、次元$m$のベクトル空間$V_m$の$n$個のベクトル$x_1, \cdots, x_n$についての、ある基底$B$に関する座標成分を表す縦ベクトルとみると、
$m$次の正則行列を左から掛けた$A'=SA$の列ベクトルは、別の基底に関する$x_1, \cdots, x_n$の座標成分を表す。

よって、$S$は正則なので、$\{x_1, \cdots, x_n\}$ の線形独立・線形従属性は、$A'$のそれを調べても可能である。

## 部分空間と補空間

**部分空間**

ベクトル空間$V$の部分集合$U$が、それ自身が一つのベクトル空間になっているとき、$U$を$V$の部分空間であるという。
例えばベクトルの集合$X \in V$に対して、
$$
\{y \in V \mid y=\lambda_1 x_1 + \cdots + \lambda_x x_m ; x_1 \cdots x_m \in X ; \lambda_1, \cdots \lambda_m \in K\}
$$
は$V$の部分空間であり、特にこれは$X$が張る部分空間という。

**補空間**

ベクトル空間$V$が二つの部分空間$U, W$の直和（共通部分が空である集合の合併）$V=U \oplus W$であるとき、$W$は$U$の補空間であるという。
それぞれが写像する空間の次元について考えれば、$\operatorname{dim} W = n - \operatorname{dim} U$は明らかである。

$U$が$V$の部分空間であるとして、

$$x - y \in U$$

であるような$x, y \in V$を考えよう。
これを満たす$x, y$によって定義される関係$x \sim y$は**同値関係**である。
この関係$\sim$によって$V$は同値類に分けられ、同値類の間に自然に和とスカラー倍を定義したものを、$V$の$U$による商空間と呼び、$V / U$と記す。

**同値関係について**

参考
* [関係、同値関係、同値類、商ベクトル空間](https://motochans.blogspot.com/2016/05/blog-post.html)

集合$S$について、「$(a, b) \in S$に関係をいれる」という動作を定義しておく。

直積集合$S\times S=\{(a, b)|a, b \in S\}$の部分集合$A$を決めて、$(a, b) \in A$であるとき、$a$と$b$に関係がある、という。
このとき、$S$の関係が同値関係であることを、次の条件を満たすこととして定義する。

$a, b, c \in S$ に対して、
* $(a, a) \in A$である。
* $(a, b) \in A$ならば$(b, a) \in A$である。
* $(a, b) \in A$かつ$(b, c) \in A$ならば $(a, c)\in A$である。

特に、集合$S$にこのような$A$を決めることを、$S$に同値関係を入れる、という。

上で定義した 
$$x - y \in U \Longleftrightarrow x\sim y$$
が同値関係であることを証明してみます。
商空間の関係に相当する集合を$A = \{(v, u) \in V \times V\mid v - u \in U\}$とします。

1. $U$はベクトル空間であるから、$v-v = 0 \in U$なので、$(v, v) \in A$
2. $v - u \in U$ ならば、$U$はベクトル空間なので、その$-1$倍である$u - v$も$U$に入る。よって、$(u, v) \in A$
3. $v - u \in U$かつ$u - w \in U$ならば、$U$はベクトル空間なので、$(v-u) + (u-w) \in U$である。よって、$(v, w) \in A$

## 線形写像

$U, V$が$K$上のベクトル空間である時、$V$から$U$への写像

$$A: V \to U$$

が線形性の条件
1. $A(x+y)=A(x) + A(y)$
2. $A(\lambda x) = \lambda A(x)$

を満足する時、$A$を$V$から$U$への線形写像という。

**線形写像の行列表現**

$V_n$、$U_m$にそれぞれ基底$B=\{e_1, \cdots, e_n\}$、$\tilde{B}=\{\tilde{e}_1, \cdots, \tilde{e}_n\}$を選んで固定しよう。
このとき、$A(e_j) \in U_m$であるから、

$$A(e_j)=\sum^m_{i=1}a_{ij} \tilde{e}_i \in U_m$$

と書ける。すると、$A$の線形性から、$V_n$の任意のベクトル

$$x = \sum^n_{j=1}x_{j} {e}_j \in V_n$$

の$A$による像は

$$A(x) = \sum^m_{i=1}(\sum^n_{j=1}x_{j} {e}_j)\tilde{e}_i \in U_m$$
なので、$A(x)=\sum^m_{i=1}\tilde{x}_i \tilde{e}_i$と比較すると、$\tilde{x}_i=\sum^n_{j=1}a_{ij}x_j$である。

この$a_{ij}$を並べることで、線形写像$A$は行列で表現できることに注意。


$l$次元、$m$次元、$n$次元のベクトル空間$U_l$、$V_m$、$W_n$の間の２つの線形写像　

$$B: W_n \to V_m, \; A: V_m \to U_l$$

の合成

$$C=A\bullet B : W_n \to U_l$$

の行列表現は、$l\times m$行列$A$と$m\times n$行列$B$の積である$l\times n$行列$C=A\bullet B$であることも明らかである。

**同値変形**

線形写像$A: V_n \to U_m$を考えよう。
このとき、$A$は$V_n$の基底$B=\{e_1,\cdots, e_n\}$と$U_m$の基底$\tilde{B}=\{\tilde{e}_1, \cdots, \tilde{e}_m\}$
を使って、
$$A(x) = \sum^m_{i=1}(\sum^n_{j=1}x_{j} {e}_j)\tilde{e}_i \in U_m$$

と書けるとする（このような書き方ができるとき、$A$は$B$と$\tilde{B}$に関する表現である、というらしい。）。

一方で、$V_n$の別の基底
$B'=\{e_1',\cdots, e_n'\}$と$U_m$の別の基底$\tilde{B}'=\{\tilde{e}_1', \cdots, \tilde{e}_m'\}$
に対する表現行列$A'$を考えてみよう。

このような$B'$と$\tilde{B}'$が、$n$次と$m$次の正則行列$T=[t_{ik}]$、$S=[s_{jl}]$によって次のように書けるとする。

$$
\begin{aligned}
[e_1',\cdots, e_n']&= [e_1,\cdots, e_n]T,\\
[\tilde{e}_1, \cdots, \tilde{e}_m] &= [\tilde{e}_1', \cdots, \tilde{e}_m']S
\end{aligned}
$$

このとき、

$$
\tilde{x} = [\tilde{e}_1, \cdots \tilde{e}_m]^T [\tilde{x}_1, \cdots, \tilde{x}_m] 
= [\tilde{e}_1, \cdots \tilde{e}_m]^T A[{x}_1, \cdots, {x}_m] 
= ([\tilde{e}_1', \cdots \tilde{e}_m']S^{-1})^T AT[{x}_1', \cdots, {x}_m'] 
= [\tilde{e}_1', \cdots \tilde{e}_m'][\tilde{x}_1', \cdots, \tilde{x}_m'] 
$$

であるから、表現行列$A'=S^{-1}AT$である。　
この形の変換を同値変換と呼ぶ。

**相似変換**

特に$A: V_n \to V_n$であるとき、$A'=S^{-1}AS$の形でかける。このような変換は相似変換と呼ぶ。
