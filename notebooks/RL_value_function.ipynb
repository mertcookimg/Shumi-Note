# 価値関数と凸性について

参考：[Reinforcement Learning: Theory and Algorithms](https://rltheorybook.github.io/)の１１章

強化学習で重要なのが状態価値関数です．方策勾配法などは一般にこの関数を最大化することを目指しますが，価値関数は一般に凸ではありません．


## 凸ではない価値関数

次の定理が成立します．

---

補題：$V^{\pi^\theta}$についての最適化問題が，direct parameterizationでもsoftmax parameterizationのどちらについても，$\theta$について凹ではないMDPが存在する．

---

次のMDPを考えてみます：

![nonconvex](figs/nonconvex-MDP.png)

このMDPでは，

$$V^\pi(s_1)=\pi(a_2|s_1)\pi(a_1|s_2) \cdot r$$

であることがわかります．また，次のパラメータを考えます：

$$
\theta^{(1)}=(\log 1, \log 3, \log 3, \log 1), \quad \theta^{(2)}=(-\log 1,-\log 3,-\log 3,-\log 1)
$$

すると，Softmax parameterizationでは

$$
\begin{aligned}
& \pi^{(1)}\left(a_2 \mid s_1\right)=\frac{3}{4} ; \quad \pi^{(1)}\left(a_1 \mid s_2\right)=\frac{3}{4} ; \quad V^{(1)}\left(s_1\right)=\frac{9}{16} r \\
& \pi^{(2)}\left(a_2 \mid s_1\right)=\frac{1}{4} ; \quad \pi^{(2)}\left(a_1 \mid s_2\right)=\frac{1}{4} ; \quad V^{(2)}\left(s_1\right)=\frac{1}{16} r . \\
&
\end{aligned}
$$

になります．このとき，$\theta^{(\mathrm{mid})}=\frac{\theta^{(1)}+\theta^{(2)}}{2}$とすれば，

$$
\pi^{(\mathrm{mid})}\left(a_2 \mid s_1\right)=\frac{1}{2} ; \quad \pi^{(\mathrm{mid})}\left(a_1 \mid s_2\right)=\frac{1}{2} ; \quad V^{(\mathrm{mid})}\left(s_1\right)=\frac{1}{4} r
$$

であり，

$$
V^{(1)}\left(s_1\right)+V^{(2)}\left(s_1\right)>2 V^{(\mathrm{mid})}\left(s_1\right),
$$

なので，このMDPでの価値関数は$\pi$について凹ではありません．
